```python
def OrderBook_A(symbol,limit):
    exchange = ccxt.bitso()
    inicio=time.time()
    data_inicio_A=strftime('%Y-%m-%d %H:%M:%S', time.localtime(time.time()))
    tiempo=0
    while tiempo<=6:
        bi_btc_ob = exchange.fetch_order_book(symbol,limit=limit)
        bi_btc_ob_time=bi_btc_ob["datetime"]
        bi_btc_ob_ask=pd.DataFrame(bi_btc_ob["asks"],columns=["Ask","AVolume"])
        bi_btc_ob_bid=pd.DataFrame(bi_btc_ob["bids"],columns=["Bid","BVolume"])
        data1 = pd.concat([bi_btc_ob_ask,  bi_btc_ob_bid],axis=1)
        data1['Index'] = data1.index 
        data1["Exchange"]=exchange
        data1["Ticker"]=symbol
        data1["Timestamp"]=bi_btc_ob_time
        data1['Vwap Ask'] =(np.cumsum(data1.Ask*data1.AVolume)/np.cumsum(data1.AVolume))
        data1['Vwap Bid'] =(np.cumsum(data1.Bid*data1.BVolume)/np.cumsum(data1.BVolume))
        data1["Total Volume"]=(data1.AVolume+data1.BVolume)
        data1["Midprice"]=((data1.Ask+data1.Bid)/2)
        data1["Spread"]=(data1.Bid-data1.Ask)
        SA=(-(((data1.Bid+data1.Ask)/2)/2)**2)
        data1["Roll Spread"]=(np.sqrt(-SA))*2
        tiempo=int(time.time()-inicio)/60
        DATA1=pd.DataFrame(data1)
        DATA1.to_csv("files\DATA1.csv",index=False,header=False, mode="a") 
        time.sleep(60)
    return data1

data_inicio_A=strftime('%Y-%m-%d %H:%M:%S', time.localtime(time.time()))
limit=50
data_A=OrderBook_A('BTC/USDT',limit)
data_A=pd.DataFrame(data_A[:limit])
pd.DataFrame(data_A[:limit])
```
